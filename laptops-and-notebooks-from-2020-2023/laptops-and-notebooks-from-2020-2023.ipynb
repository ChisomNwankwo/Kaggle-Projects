### 7. Which storage type is predominant?

### 8. What percentage of the laptops have bluetooths and webcam?

```python
with_bluetooth_webcam = laptops[(laptops['has_bluetooth'] == True) & (laptops['has_webcam'] == True)]
total_laptops = laptops.shape[0]
(with_bluetooth_webcam.shape[0] / total_laptops) * 100
```

So a good majority of the laptops have both bluetooth and webcams

### 9. what percentage of laptops have webcams, but no bluetooth?

```python
no_bluetooth_with_webcam = laptops[(laptops['has_bluetooth'] == False) & (laptops['has_webcam'] == True)]
total_laptops = laptops.shape[0]
(no_bluetooth_with_webcam.shape[0] / total_laptops) * 100
```

### 10. What percentage of laptops have no webcams and no bluetooths?

```python
no_bluetooth_no_webcam = laptops[(laptops['has_bluetooth'] == False) & (laptops['has_webcam'] == False)]
total_laptops = laptops.shape[0]
(no_bluetooth_no_webcam.shape[0] / total_laptops) * 100
```

### 10. What percentage of laptops are Apple products?

```python
apple_products = laptops[laptops['operating_system'].str.contains('mac', case=False)]
(apple_products.shape[0]/laptops.shape[0]) * 100
```
